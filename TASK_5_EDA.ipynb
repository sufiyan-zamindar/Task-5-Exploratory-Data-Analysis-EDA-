```python
# Task 5: Exploratory Data Analysis (EDA) - Titanic Dataset

# Import necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings 
warnings.filterwarnings("ignore")

# Load the Titanic dataset
df = sns.load_dataset('titanic')

# View first few rows
df.head()

# Basic Information about the dataset
print("\nDataset Info:")
df.info()

# Summary statistics
print("\nSummary Statistics:")
df.describe()

# Checking missing values
print("\nMissing Values:")
df.isnull().sum()

# Value counts for categorical features
print("\nValue Counts:")
for col in df.select_dtypes(include='object').columns:
    print(f"\nColumn: {col}")
    print(df[col].value_counts())

# --------------------------------
# Visual Explorations
# --------------------------------

# 1. Heatmap for missing values
plt.figure(figsize=(10,6))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title('Missing Values Heatmap')
plt.show()

# 2. Pairplot (selected features)
sns.pairplot(df[['survived', 'age', 'fare', 'pclass']], hue='survived')
plt.suptitle('Pairplot of Selected Features', y=1.02)
plt.show()

# 3. Correlation heatmap
plt.figure(figsize=(8,6))
numeric_df = df.select_dtypes(include=['float64', 'int64'])  # Select only numeric columns
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()


# 4. Histogram of Age
plt.figure(figsize=(8,5))
sns.histplot(df['age'].dropna(), kde=True, bins=30)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()

# 5. Boxplot of Fare by Survival
plt.figure(figsize=(8,5))
sns.boxplot(x='survived', y='fare', data=df)
plt.title('Fare vs Survival')
plt.xlabel('Survived (0=No, 1=Yes)')
plt.ylabel('Fare')
plt.show()

# 6. Countplot of Class
plt.figure(figsize=(8,5))
sns.countplot(x='pclass', data=df, palette='pastel')
plt.title('Passenger Class Distribution')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.show()

# 7. Survival rate by Gender
plt.figure(figsize=(8,5))
sns.barplot(x='sex', y='survived', data=df)
plt.title('Survival Rate by Gender')
plt.xlabel('Gender')
plt.ylabel('Survival Rate')
plt.show()

# --------------------------------
# Observations
# --------------------------------

# Observations for each plot:
print("\nObservations:")

print("""
1. Missing Values Heatmap:
   - 'Age', 'Cabin', and 'Embarked' columns have missing values.
   
2. Pairplot:
   - Survived passengers tend to have lower 'pclass' (better class) and higher 'fare'.

3. Correlation Heatmap:
   - 'Fare' and 'Pclass' show significant correlation with survival.

4. Age Distribution:
   - Most passengers are between 20-40 years of age.

5. Fare vs Survival:
   - Passengers who survived generally paid higher fares.

6. Passenger Class Distribution:
   - Majority of passengers are from class 3 (lowest class).

7. Survival Rate by Gender:
   - Females had a much higher survival rate compared to males.
""")

# --------------------------------
# Summary of Findings
# --------------------------------

print("\nSummary of Findings:")

print("""
- Younger passengers and those who paid higher fares were more likely to survive.
- Gender played a significant role in survival; females had higher chances.
- First-class passengers had higher survival rates compared to second and third-class passengers.
- Missing data, particularly in 'Age' and 'Cabin', needs careful handling for further modeling.
""")
```
